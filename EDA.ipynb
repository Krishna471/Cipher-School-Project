```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Load the dataset
df = pd.read_csv('Mall_Customers.csv')
print(df.head())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Encode categorical variables
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
print(df.head())
```

```python
# Save the cleaned dataset
df.to_csv('cleaned_mall_customers.csv', index=False)
```

```python
# Descriptive statistics
descriptive_stats = df.describe()
print(descriptive_stats)
```

```python
'''Histograms'''

# Histogram of Age
df['Age'].hist(bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Histogram of Annual Income

df['Annual Income (k$)'].hist(bins=20)
plt.title('Annual Income Distribution')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Frequency')
plt.show()

# Histogram of Spending score
df['Spending Score (1-100)'].hist(bins=20)
plt.title('Spending Score Distribution')
plt.xlabel('Spending Score')
plt.ylabel('Frequency')
plt.show()
```

```python
'''Making Scatter Plot'''

# Scatter plot annual income vs. spending score, colored by gender.

sb.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender', palette='coolwarm')
plt.title('Annual Income vs. Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
